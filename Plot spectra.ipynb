```python
%pylab inline --no-import-all
matplotlib.rcParams['image.origin'] = 'lower'
matplotlib.rcParams['image.interpolation'] = 'nearest'
import numpy as np
import matplotlib.pyplot as plt
jupiter = np.loadtxt('Jup_2AU_3x_Cahoy_formatted.txt')
sun = np.loadtxt('SunKuruczSpectrum.txt')
print(sun.shape)
fig,ax = plt.subplots()
ax.plot(jupiter[:,0],jupiter[:,1],'k',lw=2)
ax.set_xlim([0.6,1.])
minlam661 = (0.661-0.0661/2.)
maxlam661 = (0.661+0.0661/2.)
minlam721 = (0.721-0.036/2.)
maxlam721 = (0.721+0.036/2.)
minlam890 = (0.890-0.0445/2.)
maxlam890 = (0.890+0.0445/2.)
ax.plot((minlam661,maxlam661),(0.7,0.7),lw=2)
ax.plot((minlam721,maxlam721),(0.4,0.4),lw=2)
ax.plot((minlam890,maxlam890),(0.15,0.15),lw=2)
ax.set_xlabel('Wavelength (um)')
ax.set_ylabel('Geometric albedo')
ax.set_title('Jupiter spectrum at 2AU')
fig.savefig('spectra.png',dpi=300)
```
